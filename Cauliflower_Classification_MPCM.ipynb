```python
clip_bbox_cord=boxes
print(boxes)
print(arr)
mask=np.zeros((302,468,3))
for i in range(len(clip_bbox_cord)):
    x_min=int(clip_bbox_cord[i][0])
    x_max=int(clip_bbox_cord[i][2])
    y_min=int(clip_bbox_cord[i][1])
    y_max=int(clip_bbox_cord[i][3])
    mask[y_min:y_max,x_min:x_max]=1.
# only the bounding boxes will be visible Here
patch=np.multiply(arr,mask)
patch
```

```python
# color PICK values...
# signature Pixel values
v=np.array([[173. , 209. , 164.],
            [102. , 142. , 88.],
            [116. , 166. , 112.],
            [197. , 222. , 197.],
            [71. , 116. , 70.],
            [79. , 119. , 94.],
            [52. , 106. , 89. ],
            [102. , 162. , 92. ]])  #samples taken manually.

C=len(v) #number of samples
# Here distance*distance  is computed .... distance square is calculted here 
def compute_d_sq(X,v,C):
    C = len(v)
    bands=3
    dist=np.zeros((X.shape[0],X.shape[1],bands,C))
    for i in range(C):
      sub_=X-v[i]
      dist[:,:,:,i]=np.power(sub_,2)
    return np.sum(dist,axis=2)
# nu is calculated here ---
def compute_n(X,v):   
    C =len(v)
    n=np.zeros((1,C))
    dist=compute_d_sq(X,v,C)
    #w= helper(X,v,C)
    # value for every color Pick is calculated 
    for i in range (C):
        # r and c value 
        n[0,i]=(np.sum(dist[:,:,i]/(302*468)))
    return n

# mu is calculated here
def compute_u(X,n,v):
    C = len(v)
    u=np.zeros((X.shape[0],X.shape[1],C))
    dist=compute_d_sq(X,v,C)
    for i in range (C):
      u[:,:,i]=np.exp(-dist[:,:,i]/n[:,i])
    return u
```

```python
X=patch
n=compute_n(X,v)

print(n)
# mu values
u_new=compute_u(X,n,v)
print("Mu values")
print(u_new)

arr_final=np.max(u_new,axis=-1)
arr_final=arr_final*255
print("final values")
print(arr_final)
plt.figure(figsize=(10,10))
plt.imshow(arr_final)
plt.figure(figsize=(10,10))
plt.imshow(arr_final,cmap='gray')
```
